```python
npts, num_window = int(1e6), 1001
start = time()
result = conditional_window_ranks(np.random.rand(npts), np.random.rand(npts), 
                                  num_window, endpoint_fill_value='auto')
end = time()
print("Runtime for {0} points = {1:.2f} seconds".format(npts, end-start))

print("(min, max) = ({0}, {1})".format(result.min(), result.max()))
print("(median, mean) = ({0}, {1})".format(np.median(result), result.mean()))
```
